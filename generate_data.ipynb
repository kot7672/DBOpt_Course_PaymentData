```python
import pyodbc
from random import randint, choice
from datetime import datetime, timedelta
```

```python
conn = pyodbc.connect('Driver={SQL Server};'
                      'Server=DESKTOP-ARTEM;'
                      'Database=PaymentData;'
                      'Trusted_Connection=yes;')
cursor = conn.cursor()
```

```python
def connect(server='DESKTOP-ARTEM', db='PaymentData'):
    return pyodbc.connect('Driver={SQL Server};'
                      f'Server={server};'
                      f'Database={db};'
                      'Trusted_Connection=yes;')
```

```python
def random_datetime(start, end):
    return datetime.fromtimestamp(randint(int(start.timestamp()), int(end.timestamp())))
def randstr(length):
    sample = "abcdefghijklmnopqrstuvwxyz0123456789"
    return "".join([choice(sample) for _ in range(length)])
def randname(p_type):
    if p_type == 3:
        second = ('Андрей', 'Алексей', 'Сергей', 'Валерий', 'Пётр', 'Павел', 'Александр', 'Николай', 'Владимир', 'Иосиф')
        first = ('Смирнов', 'Потапов', 'Ульянов', 'Джугашвили', 'Романов', 'Артемьев', 'Лебедев', 'Уранов', 'Васильев', 'Кондратьев', 'Суровцев', 'Игнатов')
        third = ('Романович', 'Александрович', 'Владимирович', 'Васильевич', 'Юрьевич', 'Валерьевич', 'Константиноваич', 'Михайлович')
    else:
        first = ('OOO', 'AO', 'ЗАО')
        s1 = 'АБВГДЕЖЗИКЛМНОПРСТУФЭЮ'
        second = ['Евро-норд', 'Хорайзон', 'Дайнэмик', 'Северсталь', 'Кроваткин', 'Рога и Копыта']
        second.extend(["".join([choice(s1) for _ in range(randint(2, 5))]) for x in range(10)])
        third = ['Юнайтед', 'Инвест', 'Биотех', 'Риэлт', 'Строймаш', 'Интернешнл']
    return " ".join([choice(first), choice(second), choice(third)])

def translit(s):
    new_string = s
    d = {}
    d['а'] = 'a'
    d['б'] = 'b'
    d['в'] = 'v'
    d['г'] = 'g'
    d['д'] = 'd'
    d['е'] = 'e'
    d['ё'] = 'yo'
    d['ж'] = 'zh'
    d['з'] = 'z'
    d['и'] = 'i'
    d['й'] = 'y'
    d['к'] = 'k'
    d['л'] = 'l'
    d['м'] = 'm'
    d['н'] = 'n'
    d['о'] = 'o'
    d['п'] = 'p'
    d['р'] = 'r'
    d['с'] = 's'
    d['т'] = 't'
    d['у'] = 'u'
    d['ф'] = 'f'
    d['х'] = 'h'
    d['ц'] = 'ts'
    d['ч'] = 'ch'
    d['ш'] = 'sh'
    d['щ'] = 'sch'
    d['ъ'] = ""
    d['ы'] = 'y'
    d['ь'] = ""
    d['э'] = 'e'
    d['ю'] = 'yu'
    d['я'] = 'ya'
    for letter in d:
        new_string = new_string.replace(letter, d[letter])
        new_string = new_string.replace(letter.capitalize(), d[letter].capitalize())
    return new_string

def randemail(name):
    mail = ['mail', 'inbox', 'contact', 'support']
    subdomain = translit(name).replace(" ", "").lower()
    domains = ['ru', 'net', 'com', 'su', 'cc', 'org']
    return "%s@%s.%s" % (choice(mail), subdomain, choice(domains))

def randphone():
    return "".join(["+7", str(randint(8000000000, 9999999999))])

def randaddress():
    streets = ['Кузнечная', 'Ленина', 'Владимирская', 'Цветочная', 'Луначарского', 'Космонавтов', 'Строителей',
               'Конноармейская', 'Литейная', 'Воздухоплавательная', 'Геологическая', 'Архитекторов', 'Паровая',
               'Серевная', 'Южная', 'Западная', 'Восточная', 'Первооткрывателей', 'Сказочная', 'Металлургов',
               'Конюшенная', 'Ананасовая', 'Ткачей', 'Ореховая', 'Каменщиков', 'Шоколадная', 'Университетская']
    return f"{choice(streets)} ул., {randint(1, 500)}"

def randproject():
    first = ["Строительство", "Проектирование", "Создание", "Разработка", "Введение в эксплуатацию", 
             "Усовершенствование", "Реставрация", "Улучшение"]
    second = ["большого", "новейшего", "совершенного", "вечного", "универсального", "инновационного"]
    third = ["транспорта", "коллайдера", "коммуникатора", "моста", "двигателя", "космолёта", "электромобиля",
             "экскаватора", "портала", "омолаживателя", "электрогенератора", "компьютера"]
    return f"{choice(first)} {choice(second)} {choice(third)}"
```

```python
# PaymentParticipant
def fill_payment_participant(number):
    for _ in range(number):
        # balance = (randint(0, 100), randint(0, 100), randint(0, 100))
        active_from = random_datetime(datetime(2000, 1, 1), datetime.now())
        inactive_from = random_datetime(active_from, datetime.now())
        o_type = randint(0,4)
        name = randname(o_type)
        cursor.execute("""insert into [dbo].[PaymentParticipant] (
           [Oid]
          ,[Balance]
          ,[Name]
          ,[OptimisticLockField]
          ,[GCRecord]
          ,[ObjectType]
          ,[ActiveFrom]
          ,[InactiveFrom]
          ,[BankDetails]
          ,[Balance2]
          ,[Balance3]) 
          values (NEWID(), %i, '%s', %i, Null, %i, '%s', '%s', '%s', %i, %i)""" % (
            0, name, randint(1, 3), o_type, active_from.isoformat(), 
            inactive_from.isoformat(), randstr(50), 0, 0))
    conn.commit()
```

```python
# Supplier
def fill_supplier(number):
    participants = {tuple(x) for x in cursor.execute("select Oid, Name from [dbo].[PaymentParticipant]")}
    for _ in range(number):
        oid, name = participants.pop()
        contact = randemail(name)
        cursor.execute("""insert into [dbo].[Supplier] 
        ([Oid]
          ,[Contact]
          ,[ProfitByMaterialAsPayer]
          ,[ProfitByMaterialAsPayee]
          ,[CostByMaterialAsPayer])
          values ('%s', '%s', %i, %i, %i)""" % (oid, contact, randint(0, 1), randint(0, 1), randint(0, 1)))
    conn.commit()
```

```python
# Employee
def fill_employee(n_heads, n_employee):
    participants = {tuple(x) for x in cursor.execute("select Oid, Name from [dbo].[PaymentParticipant] where ObjectType = 3")}
    heads = []
    for _ in range(n_heads):
        oid, name = participants.pop()
        heads.append(oid)
        first, second, third = name.split(" ")
        q = """insert into [dbo].[Employee] ([Oid]
          ,[BusyUntil]
          ,[SecondName]
          ,[Stuff]
          ,[HourPrice]
          ,[Patronymic]
          ,[PlanfixId]
          ,[Head]
          ,[PlanfixMoneyRequestTask])
          values ('%s', '%s', '%s', Null, %i, '%s', Null, Null, Null)""" % (
            oid, random_datetime(datetime(2019, 5, 25), datetime(2019, 7, 15)).isoformat(), second, 1500, third)
        cursor.execute(q)

    for _ in range(n_employee):
        oid, name = participants.pop()
        head = choice(heads)
        first, second, third = name.split(" ")
        q = """insert into [dbo].[Employee] ([Oid]
          ,[BusyUntil]
          ,[SecondName]
          ,[Stuff]
          ,[HourPrice]
          ,[Patronymic]
          ,[PlanfixId]
          ,[Head]
          ,[PlanfixMoneyRequestTask])
          values ('%s', '%s', '%s', Null, %i, '%s', Null, '%s', Null)""" % (
            oid, random_datetime(datetime(2019, 5, 25), datetime(2019, 7, 15)).isoformat(), second, 
            randint(450, 1250), third, head)
        cursor.execute(q)
    conn.commit()
```

```python
# Client
def fill_client(number):
    clients = [x for x in cursor.execute("select Oid, Name from [dbo].[PaymentParticipant]")]
    for _ in range(number):
        client = choice(clients)
        oid = client[0]
        first, second, third = client[1].split(" ")
        phone = randphone()
        q = """insert into [dbo].[Client] (
        [Oid]
          ,[FirstName]
          ,[SecondName]
          ,[Phone])
          values ('%s', '%s', '%s', '%s')""" % (
        oid, first, second, phone)
        cursor.execute(q)
    conn.commit()
```

```python
# Project
def fill_project(number):
    employee = [tuple(x) for x in cursor.execute("select Oid, Head from [dbo].[Employee] where Head is not Null")]
    clients = [x[0] for x in cursor.execute("select Oid from [dbo].[Client]")]
    for _ in range(number):
        foreman, manager = choice(employee)
        name = randproject()
        address = randaddress()
        client = choice(clients)
        start_date = random_datetime(datetime(2019, 3, 1), datetime(2020, 5, 1))
        finish_date = random_datetime(start_date + timedelta(180), start_date + timedelta(3650))
        now = datetime.now()
        if now >= finish_date:
            status = 2
        elif now >= start_date:
            status = 1
        else:
            status = 0
        start_date = start_date.isoformat()
        finish_date = finish_date.isoformat()
        q = """insert into [dbo].[Project] ([Oid]
          ,[Name]
          ,[Address]
          ,[Client]
          ,[Manager]
          ,[Foreman]
          ,[OptimisticLockField]
          ,[GCRecord]
          ,[Balance]
          ,[BalanceByMaterial]
          ,[BalanceByWork]
          ,[PlaningStartDate]
          ,[Status]
          ,[FinishDate]
          ,[Area]
          ,[WorkPriceRate]
          ,[WorkersPriceRate]
          ,[RemainderTheAdvance]
          ,[PlanfixWorkTask]
          ,[PlanfixChangeRequestTask]
          ,[UseAnalytics]) values (
          NEWID(), '%s', '%s', '%s', '%s', '%s', %i, Null, 0, 0, 0, '%s', %i, '%s', Null, Null, Null, Null, Null, Null, Null
          )""" % (
            name, address, client, manager, foreman, randint(1, 3), start_date, status, finish_date)
        cursor.execute(q)
    conn.commit()
```

```python
# Cashbox
def fill_cashbox(number):
    account_types = [x[0] for x in cursor.execute("select Oid from [dbo].[AccountType]")]
    participants = {x[0] for x in cursor.execute("select Oid from [dbo].[PaymentParticipant]")}
    for _ in range(number):
        a_type = choice(account_types)
        oid = participants.pop()
        q = """insert into [dbo].[Cashbox] ([Oid], [AccountType]) values ('%s', '%s')""" % (
        oid, a_type)
        cursor.execute(q)
    conn.commit()
```

```python
def fill_payment(number):
    categories = [x[0] for x in cursor.execute("select Oid from [dbo].[PaymentCategory]")]
    participants = {x[0] for x in cursor.execute("select Oid from [dbo].[PaymentParticipant]")}
    projects = [x[0] for x in cursor.execute("select Oid from [dbo].[Project]")]
    if len(participants) < 2:
        return
    for _ in range(number):
        category = choice(categories)
        project = choice(projects)
        payee = participants.pop()
        payer = participants.pop()
        amount = randint(10000, 1000000)
        date = random_datetime(datetime(2017, 1, 1), datetime.now()).isoformat()
        cr_date = datetime.fromtimestamp(int(datetime.now().timestamp())).isoformat()
        q = """insert into [dbo].[Payment] (
        [Oid]
          ,[Amount]
          ,[Category]
          ,[Project]
          ,[Justification]
          ,[Comment]
          ,[Date]
          ,[Payer]
          ,[Payee]
          ,[OptimisticLockField]
          ,[GCRecord]
          ,[CreateDate]
          ,[CheckNumber]
          ,[IsAuthorized]
          ,[Number]) 
          values (NEWID(), %i, '%s', '%s', Null, Null, '%s', '%s', '%s', %i, Null, '%s', Null, Null, Null)
        """ % (amount, category, project, date, payer, payee, randint(1, 3), cr_date)
        cursor.execute(q)
    conn.commit()
```

```python
def fill_bank(number):
    a_types = [x[0] for x in cursor.execute("select Oid from [dbo].[AccountType]")]
    participants = {x[0] for x in cursor.execute("select Oid from [dbo].[PaymentParticipant]")}
    for _ in range(number):
        a_type = choice(a_types)
        oid = participants.pop()
        q = """insert into [dbo].[Bank] ([Oid], [AccountType]) 
        values ('%s', '%s')""" % (oid, a_type)
    cursor.execute(q)
    conn.commit()
```

```python
fill_payment_participant(3000)
fill_supplier(500)
fill_employee(25, 175)
fill_client(100)
fill_project(500)
fill_cashbox(700)
fill_payment(1000)
fill_bank(100)
```
